```python
import numpy as np

class SimpleLinearRegression:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.m = 0  # slope
        self.c = 0  # intercept (bias)
        self.lr = learning_rate
        self.iterations = iterations

    # 1- Hypothesis Function
    # Formula: Ŷ = mx + c
    def predict(self, X):
        """predicted values (Hypothesis)."""
        return self.m * X + self.c

    # 2- Cost Function (Mean Squared Error)
    # Formula: j(m,c) = 1/2n * Σ(y - (mx + c))^2

    def compute_cost(self, X, y):
        n = len(y)
        predictions = self.predict(X)
        cost = (1 / (2 * n)) * np.sum((y - predictions) ** 2)
        return cost

    # 3- Optimization Function (Gradient Descent)
    # Formulas:
    # ∂j/∂m = -(1/n) * Σ(X * (y - y_predicted))
    # ∂j/∂c = -(1/n) * Σ(y - y_predicted)

    def fit(self, X, y):
        n = len(y)

        for i in range(self.iterations):
            y_pred = self.predict(X)

            # Compute gradients
            dm = -(1 / n) * np.sum(X * (y - y_pred))
            dc = -(1 / n) * np.sum(y - y_pred)

            # Update parameters (m := m - α * ∂j/∂m)
            self.m = self.m - self.lr * dm
            self.c = self.c - self.lr * dc

            # Optional: Print cost every 100 steps to monitor minimization
            if i % 100 == 0:
                print(f"Iteration {i}: Cost {self.compute_cost(X, y)}")



if __name__ == "__main__":
    # Sample Data
    X = np.array([1, 2, 3, 4, 5])
    y = np.array([5, 7, 9, 11, 13])

    model = SimpleLinearRegression(learning_rate=0.01, iterations=1000)
    model.fit(X, y)

    print("\n Final Report")
    print(f"Optimal m (Slope): {model.m}")
    print(f"Optimal c (Intercept): {model.c}")
    print(f"Minimized Cost: {model.compute_cost(X, y)}")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1. Sample Data
# Note: X must be a 2D array (reshape is used for single-feature data)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([5, 7, 9, 11, 13])

# 2. Initialize and Train Model
# Scikit-learn uses OLS(Ordinary Least Squares), so iterations/learning_rate aren't needed
model = LinearRegression()
model.fit(X, y)

# 3. Predict and Calculate Cost
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)


# Final Report
print("Optimal m (Slope):", model.coef_[0])
print("Optimal c (Intercept):", model.intercept_)
print("Mean Squared Error:", mse)

```
